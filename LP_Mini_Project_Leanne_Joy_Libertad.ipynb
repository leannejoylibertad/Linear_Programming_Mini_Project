## Analytics Domain: Marketing Analytics

Marketing Analytics involves using data and quantitative methods to measure, manage, and optimize marketing performance. One of its core challenges is **media mix optimization** — determining how to allocate a limited advertising budget across multiple channels (TV, social media, radio, print) to achieve the greatest possible reach or return. Linear Programming is a natural fit for this problem because the budget is finite, each channel has defined costs and reach, and real-world business rules create clear mathematical constraints.

## Common Optimization Problem

> **"How should a company split its monthly advertising budget across TV, social media, radio, and print to maximize total audience reach without overspending on any single channel?"**

**SCENARIO**

You are a marketing analyst at **BrandBoost Philippines**, a mid-sized marketing agency handling the campaign of **FreshSip** — a local juice drink brand launching a new product line. The client has approved a monthly advertising budget of **₱500,000** and wants to run ads across four channels available in the Philippine market.

### Available Advertising Channels and Estimated Reach:

| Channel | Variable | Cost per 1,000 Impressions (CPM) | Estimated Reach per ₱1 Spent (impressions) | Audience Type |
| ------- | -------- | -------------------------------- | ------------------------------------------- | ------------- |
| TV (National) | x₁ | ₱2,000 per spot | 0.80 impressions/peso | Broad, mass market |
| Social Media (Meta/TikTok) | x₂ | ₱500 per campaign set | 2.50 impressions/peso | Youth, 18–35 |
| Radio | x₃ | ₱800 per spot | 1.20 impressions/peso | Commuters, provincial |
| Print (Broadsheet/Tabloid) | x₄ | ₱1,500 per ad | 0.50 impressions/peso | Adults, 30–55 |

### FreshSip's Campaign Rules (Constraints):
- Total spending must **not exceed ₱500,000**
- TV must receive **at least ₱80,000** (brand visibility requirement from client)
- Social Media must receive **at least ₱60,000** (digital presence requirement)
- Print must **not exceed ₱100,000** (lowest reach channel, client wants to limit it)
- **No single channel** can receive more than ₱250,000 (channel balance rule)
- All allocations must be **non-negative** (cannot unadvertise)


### "How should we allocate the ₱500,000 campaign budget across TV, social media, radio, and print to get the highest possible total audience reach — while following all of FreshSip's campaign rules?"

## A. Objective Function

We want to **maximize** the total estimated audience reach (in impressions):

**Maximize Z = 0.80·x₁ + 2.50·x₂ + 1.20·x₃ + 0.50·x₄**

Where Z = total impressions delivered, and the coefficients represent impressions earned per peso spent on each channel.

---

## B. Decision Variables

| Variable | Channel | Description | Hypothetical Range |
| -------- | ------- | ----------- | ------------------ |
| x₁ | TV (National) | Pesos allocated to TV spots | ₱80,000 – ₱250,000 |
| x₂ | Social Media | Pesos allocated to Meta/TikTok campaigns | ₱60,000 – ₱250,000 |
| x₃ | Radio | Pesos allocated to radio spots | ₱0 – ₱250,000 |
| x₄ | Print | Pesos allocated to broadsheet/tabloid ads | ₱0 – ₱100,000 |

These are the values LP will determine — the exact peso amount to allocate to each advertising channel.

---

## C. Constraints

| # | Constraint | Mathematical Form |
| - | ---------- | ----------------- |
| 1 | Total Budget | x₁ + x₂ + x₃ + x₄ ≤ 500,000 |
| 2 | Minimum TV Spend | x₁ ≥ 80,000 |
| 3 | Minimum Social Media Spend | x₂ ≥ 60,000 |
| 4 | Print Budget Cap | x₄ ≤ 100,000 |
| 5a | Channel Balance (TV) | x₁ ≤ 250,000 |
| 5b | Channel Balance (Social Media) | x₂ ≤ 250,000 |
| 5c | Channel Balance (Radio) | x₃ ≤ 250,000 |
| 5d | Channel Balance (Print) | x₄ ≤ 250,000 |
| 6 | Non-Negativity | x₁, x₂, x₃, x₄ ≥ 0 |

```python
# Load the needed library and functions
from pulp import LpMaximize, LpProblem, LpVariable, LpStatus, value
```

```python
# Create the LP model
model = LpProblem("Maximize_Advertising_Reach", LpMaximize)
```

```python
model
```

```python
# Identify the Decision Variables
# x1 = pesos allocated to TV (National)
# x2 = pesos allocated to Social Media (Meta/TikTok)
# x3 = pesos allocated to Radio
# x4 = pesos allocated to Print (Broadsheet/Tabloid)
x1 = LpVariable('x1_TV',           lowBound=0)
x2 = LpVariable('x2_Social_Media', lowBound=0)
x3 = LpVariable('x3_Radio',        lowBound=0)
x4 = LpVariable('x4_Print',        lowBound=0)
```

```python
# Define the model objective: maximize total audience reach (impressions)
# Reach coefficients = impressions earned per peso spent on each channel
model += (0.80 * x1 + 2.50 * x2 + 1.20 * x3 + 0.50 * x4)
```

```python
# Define the Constraints

# C1: Total budget must not exceed PHP 500,000
model += x1 + x2 + x3 + x4 <= 500_000     # Total campaign budget

# C2: TV must receive at least PHP 80,000 (brand visibility requirement)
model += x1 >= 80_000                       # Minimum TV spend

# C3: Social Media must receive at least PHP 60,000 (digital presence requirement)
model += x2 >= 60_000                       # Minimum social media spend

# C4: Print must not exceed PHP 100,000 (lowest reach, client limits it)
model += x4 <= 100_000                      # Print budget cap

# C5: No single channel can exceed PHP 250,000 (channel balance rule)
model += x1 <= 250_000                      # Max TV
model += x2 <= 250_000                      # Max Social Media
model += x3 <= 250_000                      # Max Radio
model += x4 <= 250_000                      # Max Print
```

```python
model
```

```python
# Solve using the pulp library
model.solve()
```

```python
print("Maximum Total Audience Reach =", value(model.objective), "impressions")
```

```python
# Sharing our Results
print("Status:", LpStatus[model.status])
for var in model.variables():
    print(f"{var.name} = PHP {var.varValue:,.2f}")
print("Maximum Total Audience Reach =", f"{value(model.objective):,.0f}", "impressions")
```

## To achieve maximum audience reach, we recommend the following advertising budget allocation for FreshSip's campaign:
- **₱80,000** allocated to TV — held at minimum required for brand visibility
- **₱250,000** allocated to Social Media — maximized to channel cap (highest reach per peso at 2.50)
- **₱170,000** allocated to Radio — remaining budget directed here for strong secondary reach
- **₱0** allocated to Print — lowest reach per peso (0.50), not allocated unless required
